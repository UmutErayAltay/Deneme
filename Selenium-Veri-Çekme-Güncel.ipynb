```python
from selenium import webdriver
import time
import pandas as pd
import numpy as np
```

```python
options = webdriver.ChromeOptions()
options.add_argument( '--no-sandbox' )
options.add_argument( '--headless' )
options.add_argument( '--disable-gpu' )
options.add_argument('--disable-dve-shm-uage')
driver = webdriver.Chrome(options=options)
```

```python
def Veri_çekme(i,url):
  try:
    detaylar = []
    fiyat = []
    tikla = driver.find_element("xpath",'//*[@id="listing-search-wrapper"]/div['+str(i)+']')
    i += 1
    tikla.click()
    element = driver.find_element('xpath', '//*[@id="bilgiler"]/div/div[2]/div/div[2]/span')
    driver.execute_script("arguments[0].click();", element)
    elements = driver.find_elements("css selector", '._2HQCBI')
    fiyatlar = driver.find_elements("css selector", '._2TxNQv')
    for j in elements:
        detaylar.append(j.text)
    for j in fiyatlar:
        fiyat.append(j.text)
    elements = driver.find_elements("css selector", '._3VQ1JB')
    print(elements[0])
    detaylar.append("\n Konum\n")
    konumlar = ["Osmangazi", "Nilüfer", "Yıldırım", "Büyükorhan", "Gemlik", "Gürsu", "Harmancık", "İnegöl",
                "İznik", "Karacabey", "Keles", "Kestel", "Mudanya", "Mustafakemalpaşa","Orhaneli","Orhangazi",
                "Yenişehir"]
    for j in konumlar:
        if elements[0].text.find(j) != -1:
            detaylar.append(j+"\n")
    fiyat_sade = fiyat[1].strip()
    fiyat_sade = fiyat_sade.replace("TL","")
    fiyat_sade = fiyat_sade.replace(".","")
    fiyat_sade = fiyat_sade.replace(",","")
    for i in range(100):
        if fiyat_sade.find("arrow_downward%"+i.__str__())!=-1 or fiyat_sade.find("arrow_upward%"+i.__str__())!=-1:
            fiyat_sade = fiyat_sade.replace("arrow_downward%"+i.__str__(),"")
            fiyat_sade = fiyat_sade.replace("arrow_upward%"+i.__str__(),"")
    fiyat_sade = int(fiyat_sade)
    print(fiyat_sade)
    içerikler = []
    içerikler.append([fiyat_sade])
    det_str = listToString(detaylar)
    ayrı = det_str.split("\n")

    df = pd.DataFrame(ayrı)
    print(df)
    df_yeni = df.iloc[0:]
    print(df_yeni)

    df_yeni = df_yeni.reset_index()
    df_yeni.drop("index", axis=1, inplace=True)
    df_liste = df_yeni.values.tolist()
    a = 1
    while a <= len(df_yeni):
        içerikler.append(df_liste[a])
        a= a+2
        if a>=len(df_yeni):
          break
    df_içerikler = pd.DataFrame(içerikler).T
    df_içerikler.to_csv(r"zingat3.csv", encoding="utf-8", index=False, mode="a")
    driver.execute_script("window.history.go(-1)")
  except:
    url_in = driver.current_url
    print(i,"Hatalı")
    driver.execute_script("window.history.go(-1)")
    if url_in != url:
        driver.get(url)
    time.sleep(1)
```

```python
def listToString(s):
    # initialize an empty string
    str1 = " "
    # return string
    return (str1.join(s))
```
